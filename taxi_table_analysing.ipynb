```python
import pandas as pd
```

```python
path_to_file = '2_taxi_nyc.csv'
```

```python
taxi = pd.read_csv(path_to_file)
```

```python
taxi
```

```python
taxi.columns
```

```python
taxi = taxi.rename(columns={'pcp 01':'pcp_01', 'pcp 06':'pcp_06', 'pcp 24':'pcp_24'})
```

```python
taxi.value_counts().get('Brooklyn')
```

```python
taxi.pickups.sum()
```

```python
taxi.query('hday == "Y"').groupby('borough').agg({'pickups': 'sum'})
```

```python
taxi.groupby(['borough','hday']).agg({'pickups': 'mean'}).unstack()
```

```python
pickups_by_mon_bor = taxi.groupby(['borough', 'pickup_month']).agg({'pickups': 'sum'}).reset_index()
pickups_by_mon_bor = pickups_by_mon_bor.sort_values(by='pickups', ascending=False)
pickups_by_mon_bor.shape
```

```python
taxi.groupby(['borough', 'pickup_month']).agg({'pickups': 'sum'}).sort_values(by='pickups')
```
