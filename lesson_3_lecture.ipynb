```python
import pandas as pd

import seaborn as sns
```

```python
data = pd.read_csv('lesson_3_data_1_.csv', encoding = 'windows-1251')
data
```

```python
users_data = data[['tc', 'art_sp']]
users_data
```

```python
users_data = users_data.rename(columns={'tc': 'user_id', 'art_sp' : 'brand_info'})
users_data
```

```python
users_data['brand_name'] = users_data.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
users_data
```

```python
users_purchases = users_data.groupby('user_id', as_index=False) \
.agg({'brand_info' : 'count'}).rename(columns={'brand_info' : 'purchases'}).query('purchases >= 5')
users_purchases
```

```python
user_lovely_brand = users_data.groupby(['user_id', 'brand_name'], as_index=False) \
        .agg({'brand_info' : 'count'}).sort_values(['user_id', 'brand_info'], ascending=[False,False]) \
.groupby('user_id').head(1).rename(columns =({'brand_name' : 'lovely_brand', 'brand_info' : 'lovely_brand_purchases' })) # в первой записи пользователя будет бренд с максимальным числом покупок
```

```python
user_lovely_brand
```

```python
users_unique_brands = users_data.groupby('user_id', as_index = False) \
.agg({'brand_name' : pd.Series.nunique}).rename(columns={'brand_name' : 'unique_brands'})
```

```python
users_unique_brands
```

```python
loyality_data = users_purchases.merge(users_unique_brands, on = 'user_id')\
.merge(user_lovely_brand,  on = 'user_id')
```

```python
loyality_data['loyality_score'] =  loyality_data.lovely_brand_purchases / loyality_data.purchases 
```

```python
loyality_data
```

```python
sns.histplot(loyality_data.loyality_score)
```

```python
a = loyality_data.groupby('lovely_brand', as_index=False).agg({'user_id' : 'count','loyality_score' : 'median'})
```
